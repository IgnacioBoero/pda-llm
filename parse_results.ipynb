```python
import wandb

api = wandb.Api()

# Replace with your actual entity and project name
entity = "alelab"
project = "Safe-RLHF-PDA-OURS"

# Fetch runs
runs = api.runs(f"{entity}/{project}")

# Loop through runs and get info
for run in runs:
    print(f"Run ID: {run.id}")
    print(f"Name: {run.name}")
    print(f"State: {run.state}")
    print(f"Config: {run.config}")           # Hyperparameters
    print(f"Summary: {run.summary}")         # Final metrics
    print(f"Metrics: {run.history()}")       # All logged metrics
    break
```

```python
# remove nans
eval  = run.history()['eval/safe_sample_ratio']
eval = eval[eval.notna()]
eval
```

```python
data = []

for run in runs:
    data.append({
        "name": run.name,
        "safe_ratio": run.summary.get("eval/safe_sample_ratio"),
        "unsafe_ratio": run.summary.get("eval/unsafe_sample_ratio"),
        "lr": run.config.get("lr"),
        "lora_r": run.config.get("lora_r"),
        "lora_alpha": run.config.get("lora_alpha"),
        "lora_dropout": run.config.get("lora_dropout"),
    })


# order by safe_ratio
data = sorted(data, key=lambda x: x["safe_ratio"], reverse=True)

data
```
